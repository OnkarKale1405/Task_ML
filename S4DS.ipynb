```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.read_csv('Project.csv')
```

```python
df.drop('Unnamed: 0' , axis=1 , inplace=True)
```

```python
df.head()
```

```python
sns.scatterplot(x='Ratings out of 5 stars',y='number_of_reviews',data=df)
```

```python
df.info()
```

```python
top_5= df['manufacturer'].value_counts().head(5)

plt.figure(figsize=(10, 6))
plt.bar(range(1,6), top_5, color='yellow')
plt.xlabel('Manufacturer')
plt.ylabel('Number of Products')
plt.title('Number of Products per Manufacturer (Top 5)')
plt.xticks(range(1,6), top_5.index)
plt.tight_layout()

# Show the bar plot
plt.show()
```

```python
df1=df[['product_name','price in £']]
df1=df.groupby(['product_name'])['price in £'].sum()
df1=df1.to_frame()
df1=df1.sort_values('price in £',ascending=False)
df2=df1.head()
df1.drop_duplicates()
df2
```

```python
df2.plot(kind='bar')

plt.title('Most expensive products(Top 5)')
plt.xlabel('product_name')
plt.ylabel('price in £')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.kdeplot(data=df, x='price in £')
```

Observation : Most of the products are priced between £ 0 to £ 100
